Prepare a model for glass classification using KNN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
import warnings
warnings.filterwarnings('ignore')
```

```python
glass=pd.read_csv("/content/glass.csv")
glass.head()
```

```python
glass.info()                      ## check for null values
```

Seperating the independent and dependent variables from the given dataset

```python
x=glass.iloc[:,0:9]
y=glass.iloc[:,9]
```

Standardising the data to avoid variance among the features in the dataset

```python
scale=preprocessing.StandardScaler()
x=scale.fit_transform(x)
```

Splitting the data into train and test datasets

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=7)
```

**Building KNN model using 4 neighbors**

```python
m=KNeighborsClassifier(n_neighbors=4)
m.fit(x_train,y_train)
```

Predicting the test data using the model

```python
y_pred=m.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```
